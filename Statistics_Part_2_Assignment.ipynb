 **3.What is the significance level in hypothesis testing, and why is it important?**
 Ans.Significance level, denoted by the Greek letter α (alpha), is a critical concept in hypothesis testing. It represents the threshold for deciding whether to reject the null hypothesis based on the evidence provided by your sample data.

The significance level is the probability of rejecting the null hypothesis when it is actually true-a mistake known as a Type I error.

Common values for α are 0.05 (5%) or 0.01 (1%), meaning you accept a 5% or 1% risk, respectively, of incorrectly concluding that an effect exists when it does not.



**4.What does a P-value represent in hypothesis testing?**
Ans.A p-value is a statistical measure that helps you determine the strength of the evidence against the null hypothesis in a hypothesis test.

Definition: The p-value is the probability of obtaining test results at least as extreme as those actually observed, assuming that the null hypothesis is true.



**5.How do you interpret the P-value in hypothesis testing?**
Ans.Interpreting the P-Value in Hypothesis Testing
The p-value represents the probability of obtaining results at least as extreme as those observed in your data, assuming the null hypothesis is true.

It quantifies how compatible your sample data are with the null hypothesis.

How to Interpret the P-Value
If the p-value is less than or equal to the significance level (commonly α = 0.05):

The result is considered statistically significant.

There is strong evidence against the null hypothesis, so you reject the null hypothesis in favor of the alternative hypothesis.

Example: If p = 0.02 and α = 0.05, reject the null hypothesis; the observed effect is unlikely to have occurred by chance.

If the p-value is greater than the significance level:

The result is not statistically significant.

There is insufficient evidence to reject the null hypothesis, so you fail to reject the null hypothesis.

Example: If p = 0.12 and α = 0.05, do not reject the null hypothesis; the observed effect could plausibly be due to random chance.


**6.What are Type 1 and Type 2 errors in hypothesis testing?**
Ans.In hypothesis testing, Type I and Type II errors represent two ways you can reach an incorrect conclusion about your data.

Type I Error (False Positive)
Definition: A Type I error occurs when you reject the null hypothesis when it is actually true.

Interpretation: You conclude that there is a significant effect or difference, but in reality, there is none. This is also known as a "false positive."

Probability: The probability of making a Type I error is denoted by alpha (α), which is the significance level you set for your test (commonly 0.05 or 5%).

Example: If a medical test incorrectly indicates a patient has a disease when they do not, that is a Type I error.

Type II Error (False Negative)
Definition: A Type II error occurs when you fail to reject the null hypothesis when it is actually false.

Interpretation: You conclude that there is no significant effect or difference, but in reality, there is one. This is also known as a "false negative."

Probability: The probability of making a Type II error is denoted by beta (β).

Example: If a medical test fails to detect a disease that the patient actually has, that is a Type II error.

**8.What is the Z-test, and when is it used in hypothesis testing?**
Ans.A Z-test is a statistical hypothesis test used to determine whether there is a significant difference between sample and population means, or between the means of two populations, under the assumption that the data are approximately normally distributed and the population standard deviation is known.

When is a Z-Test Used?
A Z-test is appropriate when:

The sample size is large (typically n ≥ 30), so the Central Limit Theorem ensures the sampling distribution is approximately normal.

The population standard deviation (σ) is known.

The data are randomly sampled and independent.

The underlying population is normally distributed, or the sample size is large enough for the normal approximation to hold.




**9. How do you calculate the Z-score, and what does it represent in hypothesis testing?**
Ans.A Z-score represents how many standard deviations a particular data point is from the mean of a dataset. It's calculated by subtracting the mean from the data point and then dividing by the standard deviation. In hypothesis testing, a Z-score is used to determine the probability of observing a data point (or a sample mean) if the null hypothesis were true.
Calculating the Z-score:
Identify the data point (x) you want to standardize.
Determine the mean (μ) and standard deviation (σ) of the dataset. If you're using a sample, you'll use the sample mean (x̄) and sample standard deviation (s).
Apply the Z-score formula:
z = (x - μ) / σ: (when using population parameters)
z = (x - x̄) / s: (when using sample parameters)
What a Z-score Represents in Hypothesis Testing:
Distance from the Mean:
A Z-score indicates how many standard deviations away from the mean a data point (or sample mean) is.
Probability:
A Z-score is used to calculate a p-value, which represents the probability of observing the data (or a more extreme result) if the null hypothesis is true. A small p-value (usually less than the significance level, α) suggests strong evidence against the null hypothesis.
Decision Rule:
In hypothesis testing, the Z-score is compared to a critical value determined by the significance level. If the Z-score falls into the rejection region (beyond the critical value), the null hypothesis is rejected.



**10. What is the T-distribution, and when should it be used instead of the normal distribution?**
Ans.The t-distribution is a probability distribution similar to the normal distribution, but with heavier tails, meaning it has a greater chance of extreme values. It's used when dealing with small sample sizes (typically less than 30) or when the population standard deviation is unknown. The normal distribution is preferred when the sample size is large or the population standard deviation is known.
Here's a more detailed explanation:
What is the t-distribution?
It's also known as Student's t-distribution.
It has a bell shape like the normal distribution but is flatter and wider.
It has heavier tails than the normal distribution, meaning there's a greater probability of observing extreme values.
The specific shape of the t-distribution depends on the degrees of freedom (related to sample size).
As the degrees of freedom increase, the t-distribution approaches the normal distribution.
When to use the t-distribution instead of the normal distribution:
Small sample sizes:
When dealing with small sample sizes (generally less than 30), the t-distribution provides a more accurate estimate of population parameters.
Unknown population standard deviation:
When the population standard deviation is unknown and needs to be estimated from the sample, the t-distribution is the appropriate choice.
Estimating confidence intervals:
The t-distribution is used to construct confidence intervals for the population mean when the sample size is small or the population standard deviation is unknown.
Hypothesis testing:
The t-distribution is also used in hypothesis testing when the sample size is small or the population standard deviation is unknown.
In essence:
If you have a large sample size and know the population standard deviation, use the normal distribution.
If you have a small sample size or don't know the population standard deviation, use the t-distribution.

**12. What is the T-test, and how is it used in hypothesis testing?**
Ans.A t-test is a statistical hypothesis test used to determine if there's a significant difference between the means of two groups. It's particularly useful when comparing the means of two samples, or comparing a sample mean to a known population mean, especially when the population standard deviation is unknown. In hypothesis testing, the t-test helps determine if observed differences between sample means are likely due to a true difference in the population or due to random chance.
How it's used in hypothesis testing:
1. Formulate Hypotheses:
Null Hypothesis (H0): There is no significant difference between the means of the two groups or between the sample mean and the population mean.
Alternative Hypothesis (H1): There is a significant difference between the means.
2. Set Significance Level (α):
This is the probability of rejecting the null hypothesis when it is actually true. Common values are 0.05 or 0.01.
3. Calculate the Test Statistic:
The t-test calculates a t-value, which measures the difference between the sample means (or the sample mean and population mean) relative to the variability within the samples.
4. Determine Degrees of Freedom:
Degrees of freedom (df) depend on the type of t-test (e.g., independent samples t-test, paired t-test) and the sample sizes.
5. Find the Critical Value(s):
Using the t-distribution table and the significance level (α), determine the critical t-value(s) for the specified degrees of freedom.
6. Make a Decision:
If the calculated t-value is greater than the critical value(s), reject the null hypothesis. This indicates that there's a statistically significant difference between the means.
If the calculated t-value is not greater than the critical value(s), fail to reject the null hypothesis. This suggests that the observed difference may be due to random chance.
Types of T-tests:
Independent Samples T-test:
Used when comparing the means of two independent groups (e.g., comparing the heights of two different groups of students).
Paired T-test:
Used when comparing the means of two related groups (e.g., measuring the same students' heights before and after a growth intervention).
One-Sample T-test:
Used when comparing the mean of a sample to a known population mean (e.g., testing if the average height of a sample of students differs from the national average height.)




**13.What is the relationship between Z-test and T-test in hypothesis testing?**
Ans.Z-test and T-test are both statistical tools used in hypothesis testing to determine whether there is a significant difference between sample and population data. They are closely related but have key differences:
- Applicability: A Z-test is used when the sample size is large (typically n > 30) and the population variance is known. A T-test, on the other hand, is used when the sample size is small (n ≤ 30) or when the population variance is unknown.
- Distribution Used: Z-tests rely on the standard normal distribution (Z-distribution), whereas T-tests use the Student's T-distribution, which accounts for increased variability when sample sizes are small.
- Precision: Because a T-test adjusts for smaller sample sizes, it generally provides better estimates when dealing with fewer data points.
Simply put, both tests measure whether sample means significantly differ from a known population mean, but the T-test is more flexible for smaller samples.




**14.What is a confidence interval, and how is it used to interpret statistical results?**
Ans.A confidence interval (CI) is a range of values that estimates an unknown population parameter—typically the mean—based on sample data. It provides a degree of certainty about the estimate and is widely used in statistical analysis.
How It Works:
- A confidence interval consists of two boundaries, called the lower and upper limits, and is usually expressed in the form:
[ (\text{lower limit}, \text{upper limit}) ]
- It is calculated using sample statistics (like the sample mean) plus/minus a margin of error based on the standard deviation and sample size.
- The confidence level (e.g., 95%) represents the probability that the interval contains the true population parameter. A 95% confidence interval means that if we repeated the experiment many times, about 95% of the calculated intervals would contain the true mean.
How It Helps in Interpretation:
- Decision Making: If a confidence interval for a mean difference includes zero, it suggests there is no significant difference between groups.
- Estimating Population Parameters: It helps estimate unknown values, ensuring that statistical conclusions are backed by sample data.
- Assessing Uncertainty: The wider the confidence interval, the greater the uncertainty in the estimate (small samples tend to have wider intervals).



**15.What is the margin of error, and how does it affect the confidence interval?**
Ans.The margin of error (MOE) represents the amount of uncertainty in an estimate—it defines how much the sample statistic could vary from the true population value. It plays a crucial role in constructing a confidence interval.
Understanding Margin of Error:
- It accounts for sampling variability, reflecting how different a sample statistic might be from the actual population parameter.
- Mathematically, it's calculated as: [ MOE = Z \times \frac{\sigma}{\sqrt{n}} ] Where:
- ( Z ) is the critical value from a Z or T-distribution (depends on confidence level),
- ( \sigma ) is the standard deviation of the population,
- ( n ) is the sample size.
Impact on Confidence Interval:
- Determines the Width: A larger margin of error leads to a wider confidence interval, indicating more uncertainty, while a smaller MOE leads to a narrower interval, implying greater precision.
- Sample Size Influence: Increasing the sample size reduces MOE, leading to more precise estimates.
- Confidence Level Dependence: Higher confidence levels (like 99%) result in a larger MOE, widening the confidence interval to ensure greater certainty.
In simple terms, the margin of error helps define the reliability of the confidence interval.



**16. How is Bayes' Theorem used in statistics, and what is its significance?**
Ans.Bayes' Theorem is a fundamental principle in probability theory and statistics that helps update the probability of an event based on new evidence. It is widely used in various fields, including machine learning, medicine, and finance.
Formula of Bayes' Theorem
[ P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} ] Where:
- ( P(AB) ) is the posterior probability (the probability of event A given B has occurred),
- ( P(BA) ) is the likelihood (probability of event B given A),
- ( P(A) ) is the prior probability (initial probability of A before observing B),
- ( P(B) ) is the total probability of event B occurring.
How It's Used in Statistics
- Medical Diagnosis: Used to determine the likelihood of a disease given a test result.
- Spam Filtering: Helps classify emails as spam or legitimate based on prior probabilities and new data.
- Machine Learning: Plays a key role in probabilistic models, like Naïve Bayes classifiers.
- Risk Assessment: Used in financial markets and insurance to update risk probabilities.
Significance of Bayes' Theorem
- Improves decision-making by incorporating new evidence.
- Balances prior knowledge and new data, making predictions more accurate.
- Widely applicable across domains requiring probabilistic reasoning.

**18.What is the Chi-square goodness of fit test, and how is it applied?**
Ans.The Chi-square goodness of fit test is a statistical test used to determine whether a sample distribution matches an expected distribution. It is useful when analyzing categorical data and assessing how well observed frequencies align with theoretical predictions.
How It Works
This test compares observed frequencies ((O)) from a sample with expected frequencies ((E)) from a known or hypothesized distribution. The test statistic is calculated as:
[ \chi^2 = \sum \frac{(O - E)^2}{E} ]
Where:
- ( O ) = observed frequency in each category
- ( E ) = expected frequency in each category
- The sum is taken over all categories



**19.What is the F-distribution, and when is it used in hypothesis testing?**
Ans.The F-distribution is a probability distribution that arises frequently in hypothesis testing, particularly in comparing variances and analyzing variance models. It is a right-skewed distribution that only takes positive values, and its shape depends on two parameters: degrees of freedom for the numerator ((df_1)) and the denominator ((df_2)).
When is the F-Distribution Used?
- Analysis of Variance (ANOVA)
- Used to compare means across multiple groups to determine if at least one group differs significantly.
- Example: Comparing student performance across three different teaching methods.
- F-Test for Equality of Variances
- Tests whether two populations have the same variance.
- Example: Assessing if the variability in test scores differs between two schools.
- Regression Analysis
- Determines the significance of predictors in multiple regression models.
- Example: Evaluating whether marketing spend, customer engagement, and pricing significantly impact sales.
F-Statistic Formula
The test statistic for the F-test is given by:
[ F = \frac{\text{Variance of Group 1}}{\text{Variance of Group 2}} ]
If ( F ) is significantly larger than expected under the null hypothesis, we may reject (H_0), indicating that variances or means differ significantly.


**20.What is an ANOVA test, and what are its assumptions?**
Ans.ANOVA (Analysis of Variance) is a statistical test used to compare the means of multiple groups to determine if there are significant differences between them. It is particularly useful in experimental research and helps assess whether variations in data are due to real differences or random chance.
Types of ANOVA
- One-Way ANOVA – Used when comparing one independent variable across multiple groups.
Example: Testing if three different diets lead to different weight loss outcomes.
- Two-Way ANOVA – Used when comparing two independent variables and their interaction.
Example: Analyzing how both exercise level and diet affect weight loss.
Assumptions of ANOVA
For reliable results, ANOVA assumes:
- Independence of observations
- The data points should be independent from each other.
- Normality
- The dependent variable should be normally distributed in each group.
- Homogeneity of variance (Homoscedasticity)
- The variances across groups should be approximately equal (checked using Levene’s test).
- Random Sampling
- The data should be randomly collected to ensure unbiased results.



**21.What are the different types of ANOVA tests?**
Ans.ANOVA (Analysis of Variance) has several types, each designed for different statistical comparisons. Here are the main types:
1. One-Way ANOVA
- Used when comparing one independent variable with multiple groups.
- Example: Testing whether three different fertilizers lead to different plant growth rates.
2. Two-Way ANOVA
- Used when analyzing two independent variables simultaneously.
- Example: Examining how diet and exercise together impact weight loss.
3. Repeated Measures ANOVA
- Used when comparing means across multiple measurements of the same subjects.
- Example: Measuring patients' blood pressure before, during, and after a treatment.
4. MANOVA (Multivariate ANOVA)
- Extends ANOVA to compare multiple dependent variables simultaneously.
- Example: Studying how different teaching methods affect both math and science scores.
5. ANCOVA (Analysis of Covariance)
- A modified ANOVA that adjusts for confounding variables.
- Example: Testing drug effectiveness while controlling for patients' age.
Each ANOVA type serves a specific purpose, ensuring accurate comparisons across varied datasets.



**22.What is the F-test, and how does it relate to hypothesis testing?**
Ans.The F-test is a statistical test used to compare variances or determine the significance of multiple factors in a dataset. It is widely applied in hypothesis testing, especially in ANOVA (Analysis of Variance) and regression analysis.
How It Relates to Hypothesis Testing
The F-test helps evaluate whether:
- Two populations have equal variances (used in variance comparison).
- Multiple groups have significant mean differences (ANOVA).
- Regression models explain variation in the dependent variable (testing significance of predictors).
Formula for F-Test
[ F = \frac{\text{Variance of Group 1}}{\text{Variance of Group 2}} ]
- If ( F ) is significantly large, it suggests a real difference, leading to rejection of the null hypothesis ( H_0 ).
- If ( F ) is close to 1, variances are similar, and we fail to reject ( H_0 ).
Application Example
Suppose you’re comparing two teaching methods and want to check if students' performance varies significantly. The F-test can assess whether the variance in test scores is different between methods, helping decide if the teaching strategy truly impacts learning outcomes.

```python
#3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1)
    p (float): Probability of success (0 ≤ p ≤ 1)

    Returns:
    float: Probability of the given outcome
    """
    if x not in [0, 1]:
        raise ValueError("Bernoulli distribution is only defined for x = 0 or x = 1.")

    return p if x == 1 else (1 - p)

# Example usage:
p = 0.7  # Probability of success
print("P(X=1):", bernoulli_pdf(1, p))
print("P(X=0):", bernoulli_pdf(0, p))

```

```python
#4.Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters for the binomial distribution
n = 10   # Number of trials
p = 0.5  # Probability of success
size = 1000  # Number of samples

# Generate random binomial samples
data = np.random.binomial(n, p, size)

# Plot the histogram
plt.hist(data, bins=range(n+2), density=True, alpha=0.7, color='skyblue', edgecolor='black', rwidth=0.8)

# Overlay the theoretical probability mass function (PMF)
x = np.arange(0, n+1)
pmf_values = binom.pmf(x, n, p)
plt.plot(x, pmf_values, marker='o', linestyle='-', color='red', label="Theoretical PMF")

# Labels and title
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title(f"Binomial Distribution (n={n}, p={p})")
plt.xticks(range(n+1))
plt.legend()
plt.show()
```

```python
#5.Create a Poisson distribution and visualize it using Python.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define Poisson distribution parameter (lambda = mean number of events)
lambda_value = 5  # Average number of occurrences in a given interval

# Generate Poisson-distributed random samples
samples = np.random.poisson(lambda_value, 1000)

# Plot the histogram of the samples
plt.hist(samples, bins=range(0, max(samples)+2), density=True, alpha=0.7, color='skyblue', edgecolor='black', rwidth=0.8, label="Simulated Data")

# Overlay the theoretical probability mass function (PMF)
x = np.arange(0, max(samples)+1)
pmf_values = poisson.pmf(x, lambda_value)
plt.plot(x, pmf_values, marker='o', linestyle='-', color='red', label="Theoretical PMF")

# Labels and title
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.title(f"Poisson Distribution (λ={lambda_value})")
plt.xticks(x)
plt.legend()
plt.show()
```

```python
#9.Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a given dataset.

    Parameters:
    data (list or numpy array): The dataset

    Returns:
    numpy array: Z-scores of the dataset
    """
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # ddof=1 for sample standard deviation
    return (data - mean) / std_dev

def plot_z_scores(data):
    """
    Plot the Z-scores for a given dataset.

    Parameters:
    data (list or numpy array): The dataset
    """
    z_scores = calculate_z_scores(data)

    plt.figure(figsize=(8, 5))
    plt.scatter(range(len(data)), z_scores, color='skyblue', edgecolors='black')
    plt.axhline(y=0, color='red', linestyle='--', label="Mean (Z=0)")
    plt.axhline(y=2, color='green', linestyle='--', label="Z=2")
    plt.axhline(y=-2, color='green', linestyle='--', label="Z=-2")
    plt.xlabel("Index")
    plt.ylabel("Z-score")
    plt.title("Z-scores of the Dataset")
    plt.legend()
    plt.grid()
    plt.show()

# Example usage
data = np.random.normal(loc=50, scale=10, size=100)  # Generate sample data
plot_z_scores(data)
```

```python
#16.Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
import scipy.stats as stats

# Generate a sample dataset (normally distributed)
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)  # Mean=50, Std=10, Sample Size=100

# Compute sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # ddof=1 for unbiased estimate
sample_size = len(data)

# Define confidence level
confidence_level = 0.95
alpha = 1 - confidence_level
t_critical = stats.t.ppf(1 - alpha/2, df=sample_size-1)  # Two-tailed t critical value

# Calculate margin of error (MOE)
margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))

# Compute confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error
confidence_interval = (lower_bound, upper_bound)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
#24.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
#distribution.
import numpy as np
import scipy.stats as stats

def compute_skewness_kurtosis(data):
    """
    Calculate skewness and kurtosis for a given dataset.

    Parameters:
    data (numpy array): The dataset

    Returns:
    tuple: (skewness, kurtosis)
    """
    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data, fisher=True)  # Fisher=True returns excess kurtosis (normal = 0)
    return skewness, kurtosis

# Generate a dataset from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, Std=10

# Compute skewness and kurtosis
skewness, kurtosis = compute_skewness_kurtosis(data)

# Print results
print(f"Skewness: {skewness:.2f}")
print(f"Kurtosis (Excess): {kurtosis:.2f}")
```

```python
#5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2(alpha=0.05, beta=0.2, true_mean=100, sample_std=15):
    """
    Visualize Type 1 and Type 2 errors in hypothesis testing.

    Parameters:
    alpha (float): Significance level (Type 1 Error rate).
    beta (float): Probability of Type 2 Error.
    true_mean (float): True population mean.
    sample_std (float): Standard deviation of the sample.
    """
    # Compute critical value for rejecting H0
    z_critical = norm.ppf(1 - alpha)

    # Compute the threshold for accepting H0 (assuming two-tailed test)
    rejection_boundary = true_mean + z_critical * sample_std

    # Define x-axis values
    x = np.linspace(true_mean - 4 * sample_std, true_mean + 4 * sample_std, 1000)

    # Compute PDFs for Null (H0 True) and Alternative (H0 False)
    h0_distribution = norm.pdf(x, true_mean, sample_std)
    h1_distribution = norm.pdf(x, true_mean + sample_std, sample_std)

    # Plot distributions
    plt.figure(figsize=(8, 5))
    plt.plot(x, h0_distribution, label="H0: Null Hypothesis (True Mean)", color="blue")
    plt.plot(x, h1_distribution, label="H1: Alternative Hypothesis (Shifted Mean)", color="red")

    # Shade Type 1 Error (Rejecting True H0)
    plt.fill_between(x, h0_distribution, where=(x > rejection_boundary), color="purple", alpha=0.5, label="Type 1 Error (α)")

    # Shade Type 2 Error (Failing to Reject False H0)
    plt.fill_between(x, h1_distribution, where=(x < rejection_boundary), color="green", alpha=0.5, label="Type 2 Error (β)")

    # Plot critical boundary
    plt.axvline(rejection_boundary, color="black", linestyle="--", label="Critical Value")

    # Labels and title
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.title("Visualizing Type 1 & Type 2 Errors in Hypothesis Testing")
    plt.legend()
    plt.grid()
    plt.show()

# Call function to visualize errors
visualize_type1_type2()
```

```python
#6.Write a Python program to perform an independent T-test and interpret the results.
import numpy as np
import scipy.stats as stats

# Generate two independent sample datasets
np.random.seed(42)
group1 = np.random.normal(loc=75, scale=10, size=30)  # Mean=75, Std=10, Size=30
group2 = np.random.normal(loc=80, scale=12, size=30)  # Mean=80, Std=12, Size=30

# Perform independent t-test
t_statistic, p_value = stats.ttest_ind(group1, group2, equal_var=True)  # Assuming equal variance

# Print results
print(f"T-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation of results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The means of the two groups are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the two groups.")
```

```python
#7.Perform a paired sample T-test using Python and visualize the comparison results.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generate paired sample datasets (before and after intervention)
np.random.seed(42)
before = np.random.normal(loc=75, scale=10, size=30)  # Mean=75, Std=10, Size=30
after = before + np.random.normal(loc=3, scale=5, size=30)  # Increase by ~3 points on average

# Perform paired sample T-test
t_statistic, p_value = stats.ttest_rel(before, after)

# Print results
print(f"T-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The means of the two groups are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference detected.")

# Visualization: Boxplot comparison
plt.figure(figsize=(8, 5))
plt.boxplot([before, after], labels=["Before Training", "After Training"], patch_artist=True,
            boxprops=dict(facecolor="skyblue"), medianprops=dict(color="red", linewidth=2))

plt.title("Paired Sample Comparison")
plt.ylabel("Test Scores")
plt.grid()
plt.show()
```

```python
#8.Simulate data and perform both Z-test and T-test, then compare the results using Python.
import numpy as np
import scipy.stats as stats

# Generate a sample dataset
np.random.seed(42)
population_mean = 100
population_std = 15  # Known for Z-test
sample_size = 25  # Small sample to see T-test effect
sample_data = np.random.normal(population_mean, population_std, sample_size)  # Simulated sample

# Compute sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation

# Perform one-sample Z-test (Known population std)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value_z = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test

# Perform one-sample T-test (Unknown population std)
t_statistic, p_value_t = stats.ttest_1samp(sample_data, population_mean)  # Two-tailed test

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-test: Z-score = {z_score:.2f}, P-value = {p_value_z:.4f}")
print(f"T-test: T-statistic = {t_statistic:.2f}, P-value = {p_value_t:.4f}")

# Interpretation
alpha = 0.05
if p_value_z < alpha:
    print("Z-test: Reject the null hypothesis - significant difference detected.")
else:
    print("Z-test: Fail to reject the null hypothesis - no significant difference detected.")

if p_value_t < alpha:
    print("T-test: Reject the null hypothesis - significant difference detected.")
else:
    print("T-test: Fail to reject the null hypothesis - no significant difference detected.")
```

```python
#9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
import scipy.stats as stats

def compute_confidence_interval(data, confidence_level=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    data (list or numpy array): Sample dataset
    confidence_level (float): Confidence level (default = 95%)

    Returns:
    tuple: (Lower bound, Upper bound)
    """
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation
    sample_size = len(data)

    # Compute critical t-value for the given confidence level
    alpha = 1 - confidence_level
    t_critical = stats.t.ppf(1 - alpha / 2, df=sample_size - 1)  # Two-tailed test

    # Compute margin of error (MOE)
    margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))

    # Confidence Interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)  # Generate sample data
ci_lower, ci_upper = compute_confidence_interval(data)

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#10.Write a Python program to calculate the margin of error for a given confidence level using sample data.
import numpy as np
import scipy.stats as stats

def calculate_margin_of_error(data, confidence_level=0.95):
    """
    Calculate the margin of error for a sample mean.

    Parameters:
    data (list or numpy array): Sample dataset
    confidence_level (float): Confidence level (default = 95%)

    Returns:
    float: Margin of error
    """
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Use ddof=1 for unbiased estimate
    sample_size = len(data)

    # Compute critical Z-score
    alpha = 1 - confidence_level
    z_critical = stats.norm.ppf(1 - alpha / 2)  # Two-tailed test

    # Compute margin of error (MOE)
    margin_of_error = z_critical * (sample_std / np.sqrt(sample_size))

    return margin_of_error

# Example usage
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)  # Generate sample data
moe = calculate_margin_of_error(data)

print(f"Sample Mean: {np.mean(data):.2f}")
print(f"95% Margin of Error: ±{moe:.2f}")
```

```python
#11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
import numpy as np

def bayesian_inference(prior, likelihood, total_prob):
    """
    Apply Bayes' Theorem to update probability.

    Parameters:
    prior (float): Prior probability P(A)
    likelihood (float): Probability of evidence given A, P(B|A)
    total_prob (float): Total probability of evidence, P(B)

    Returns:
    float: Posterior probability P(A|B)
    """
    posterior = (likelihood * prior) / total_prob
    return posterior

# Example: Disease detection scenario
prior = 0.01  # Probability of having the disease P(A)
likelihood = 0.95  # Test accuracy P(B|A)
false_positive_rate = 0.05  # Probability of test being positive without disease P(B|¬A)
prob_no_disease = 1 - prior  # Probability of not having the disease

# Compute total probability of a positive test P(B)
total_prob = (likelihood * prior) + (false_positive_rate * prob_no_disease)

# Compute posterior probability P(A|B)
posterior = bayesian_inference(prior, likelihood, total_prob)

print(f"Probability of having the disease after a positive test: {posterior:.4f}")
```

```python
#12.Perform a Chi-square test for independence between two categorical variables in Python.
import numpy as np
import scipy.stats as stats

# Step 1: Create a contingency table (observed frequencies)
# Rows: Age Group (Young, Old)
# Columns: Drink Preference (Coffee, Tea)
observed = np.array([[30, 20],  # Young group
                     [40, 10]]) # Old group

# Step 2: Perform Chi-square test for independence
chi2_stat, p_value, dof, expected = stats.chi2_contingency(observed)

# Step 3: Print results
print(f"Chi-square Statistic: {chi2_stat:.2f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p_value:.4f}")
print(f"Expected Frequencies:\n{expected}")

# Step 4: Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant association between age group and drink preference.")
else:
    print("Fail to reject the null hypothesis: No significant association detected.")
```

```python
#13.Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
#data.
import numpy as np
import scipy.stats as stats

def compute_expected_frequencies(observed):
    """
    Compute expected frequencies for a Chi-square test.

    Parameters:
    observed (numpy array): Contingency table of observed counts.

    Returns:
    numpy array: Expected frequencies table.
    """
    row_totals = observed.sum(axis=1, keepdims=True)
    col_totals = observed.sum(axis=0, keepdims=True)
    grand_total = observed.sum()

    expected = (row_totals @ col_totals) / grand_total
    return expected

# Example: Observed data (contingency table)
observed = np.array([[30, 20],  # Group 1 preference
                     [40, 10]]) # Group 2 preference

# Compute expected frequencies
expected_frequencies = compute_expected_frequencies(observed)

print("Observed Frequencies:")
print(observed)
print("\nExpected Frequencies:")
print(expected_frequencies)
```

```python
#14.Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
import numpy as np
import scipy.stats as stats

# Observed frequencies (survey responses)
observed = np.array([45, 25, 30])  # Action, Comedy, Drama

# Expected frequencies (based on expected percentages)
total_responses = np.sum(observed)
expected_proportions = np.array([0.4, 0.3, 0.3])  # Expected distribution percentages
expected = expected_proportions * total_responses  # Compute expected frequencies

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

# Print results
print(f"Chi-square Statistic: {chi2_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: No significant deviation from the expected distribution.")
```
